```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
df = pd.read_csv("clean_dataset.csv")
df_viz = df.copy()
df_viz['target'] = df_viz['target'].map({0: 'Malignant', 1: 'Benign'})
palette = {'Malignant': '#ff4d4d', 'Benign': '#66cc66'}
plt.figure(figsize=(6, 5))
sns.countplot(
    x='target',
    hue='target',
    data=df_viz,
    palette=palette,
    edgecolor='black',
    legend=False
)
plt.title("Distribution of Classes", fontweight='bold')
plt.xlabel("Cancer Type")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_style("whitegrid")
df = pd.read_csv("clean_dataset.csv")
df_ml = df.copy()

corr = df_ml.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
plt.figure(figsize=(10, 7))
sns.heatmap(
    corr,
    mask=mask,
    cmap='coolwarm',
    cbar=True
)
plt.title("Correlation Heatmap", fontweight='bold')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
sns.set_style("whitegrid")
df = pd.read_csv("clean_dataset.csv")
df_ml = df.copy()
X = df_ml.drop('target', axis=1)
y = df_ml['target']
model = RandomForestClassifier(random_state=42)
model.fit(X, y)
importances = model.feature_importances_
feat_importances = pd.Series(importances, index=X.columns)
plt.figure(figsize=(8, 5))
feat_importances.nlargest(10).plot(
    kind='barh',
    color='#5DADE2',
    edgecolor='black'
)
plt.title("Top 10 Important Features", fontweight='bold')
plt.xlabel("Feature Importance")
plt.gca().invert_yaxis()
plt.show()
```
